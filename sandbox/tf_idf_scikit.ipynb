```python
synth_df = pd.read_csv(os.path.join(os.getcwd(), "../data/synthetic_data.csv"))
synth_df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

corpus = synth_df["text"].tolist()
metadata = synth_df[["id", "category"]].to_dict(orient = "records")

vectorizer = TfidfVectorizer()
vector = vectorizer.fit_transform(corpus)
dict(zip(vectorizer.get_feature_names_out(), vector.toarray()[0]))

tfidf_df = pd.DataFrame(
    vector.toarray(), index = metadata, columns = vectorizer.get_feature_names_out()
)

vector.toarray().shape

query_vector = vectorizer.transform(["wireless earbuds"])
query_vector.toarray().shape


similarities = cosine_similarity(X = query_vector, Y = vector) #Returns ndarray of scores

top_indices = np.argsort(similarities)[0][::-1][0:3]

top_texts = [corpus[idx] for idx in top_indices]
top_metadata = [metadata[idx] for idx in top_indices]

top_texts
top_metadata
```
